```python
import pandas as pd
import numpy as np

def clean_sales_data(file_path):

    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
        return None

    print("Original DataFrame columns:", df.columns)

    df.dropna(subset=['Product_Category', 'Order_Quantity'], inplace=True)
    print("Rows with missing 'Product_Category' or 'Order_Quantity' have been removed.")

    df['Unit_Price'] = df['Unit_Price'].astype(str).str.replace('[$,]', '', regex=True)
    df['Unit_Price'] = pd.to_numeric(df['Unit_Price'], errors='coerce')

    mean_price = df['Unit_Price'].mean()
    df['Unit_Price'].fillna(mean_price, inplace=True)
    print("Missing 'Unit_Price' values have been filled with the mean.")

    missing_revenue_mask = df['Revenue'].isnull()

    df['Order_Quantity'] = pd.to_numeric(df['Order_Quantity'], errors='coerce')
    df['Unit_Price'] = pd.to_numeric(df['Unit_Price'], errors='coerce')


    df.loc[missing_revenue_mask, 'Revenue'] = df.loc[missing_revenue_mask, 'Order_Quantity'] * df.loc[missing_revenue_mask, 'Unit_Price']
    print("Missing 'Revenue' values have been calculated.")

    print("\nCleaned Data (first 5 rows):")
    display(df.head())

    return df

if __name__ == '__main__':

    cleaned_df = clean_sales_data('/content/Bike_Sales_CSV_Lab.csv')

    if cleaned_df is not None:
        cleaned_df.to_csv('cleaned_sales_data.csv', index=False)
        print("\nCleaned data has been saved to 'cleaned_sales_data.csv'.")
```

```python
import pandas as pd
import numpy as np

# Load the dataset
df_applicants = pd.read_csv('/content/applicants.csv')

# Step 1: Drop rows with missing Python_Score
df_applicants.dropna(subset=['Python_Score'], inplace=True)

# Step 2: Replace missing Experience_Years with 0
df_applicants['Experience_Years'].fillna(0, inplace=True)

# Step 3: Filter applicants with Python_Score > 80 and Experience_Years > 2
experienced_python_candidates = df_applicants[
    (df_applicants['Python_Score'] > 80) &
    (df_applicants['Experience_Years'] > 2)
].reset_index(drop=True)

# Save the final filtered data to a new CSV file
experienced_python_candidates.to_csv('filtered_applicants.csv', index=False)
```
